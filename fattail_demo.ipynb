```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk(nsteps):
    x = 0.0
    for i in range(nsteps):
        dx = np.random.choice([-1.0, 1.0])
        x += dx
    return x

def shoot_arrow(d):
    # choose a random angle between -(pi/2) and (pi/2)
    theta = np.random.uniform(-np.pi/2, np.pi/2)

    # a triangle is formed by the angle theta and the adjacent leg of size d. The opposite leg is
    # the distance the arrow travels in the x direction. Calculate this distance.
    x = d * np.tan(theta)

    return x

TRIALS = 1000
RANDOM_WALK_STEPS = 100
DISTANCE_FROM_WALL = 5

# create an array of TRIALS random walks of 100 steps
walks = np.array([random_walk(RANDOM_WALK_STEPS) for i in range(TRIALS)])

# create an array of 
arrows = np.array([shoot_arrow(DISTANCE_FROM_WALL) for i in range(TRIALS)])

# plot the histogram of the both arrays
min_walks = walks.min()
max_walks = walks.max()

plt.hist(walks, bins=int(max_walks - min_walks), range=(min_walks, max_walks), alpha=0.5, label='Random Walks')
#plt.hist(arrows, bins=(max_walks - min_walks), range=(min_walks, max_walks), alpha=0.5, label='Arrows')
plt.legend(loc='upper right')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Walks and Arrows')
plt.show()


```
